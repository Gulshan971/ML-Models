```python
fig ,  (ax1 , ax2 ) =  plt.subplots(nrows = 2, ncols = 1 , figsize = (14 ,6))
df.Inches = pd.to_numeric(frame.Inches , errors = 'coerce')
sns.kdeplot(x = frame.Price , hue = frame.Company , ax = ax1)
sns.barplot(y = frame.Inches , x = frame.Company , data = frame , hue = frame.Company  , ax = ax2)
plt.show()
```
